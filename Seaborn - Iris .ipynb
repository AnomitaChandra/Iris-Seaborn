```python
#Countplot using seaborn

ax = sns.countplot(iris.Species,label="Count", palette = 'GnBu_d')
```

```python
#Scatterplot

ax = sns.scatterplot(x = 'Sepal_l', y = 'Sepal_w', hue = iris.Species, style = iris.Species, data = iris, palette = 'GnBu_d')
```

```python
#Linechart
x = sns.lineplot(data=iris.drop(['Species'], axis = 1), palette= 'GnBu_d')
```

```python
#Catplot - swarm
p = sns.catplot(x="Species", y="Petal_l", kind = 'swarm', data= iris, palette= 'GnBu_d')
```

```python
#Catplot - violin
#iris = pd.melt(iris,id_vars="Species",
  #                  var_name="Petal_w",
   #                 value_name='Petal_l')
p = sns.catplot(x= "Species", y="Sepal_l", kind = 'violin', hue = 'Species',data= iris,palette= 'GnBu_d')
```

```python
#Relplot
p = sns.relplot(x="Petal_l", y="Petal_w",col="Species", hue="Species",kind="line", style = 'Species', data= iris,palette= 'GnBu_d' )
```

```python
#Heatmap
f, ax = plt.subplots(figsize=(8, 6))
b = sns.heatmap(iris.corr(), annot=True, linewidths=.5, ax=ax)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

```python
#Facetgrid
g = sns.FacetGrid(iris, col='Species')
g = g.map(sns.kdeplot, 'Sepal_l')
```

```python
#Pairplot
x = sns.pairplot(iris)
```

```python
#Jointplot
a = sns.jointplot("Sepal_l", "Sepal_w", data = iris, kind = 'reg')
```
